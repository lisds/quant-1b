# Dr. Arbuthnot’s Baptism Records

In 1710, Dr. John Arbuthnot published a paper in the Royal Society's *Philosophical Transactions* called "An argument for Divine Providence, taken from the constant regularity observed in the births of both sexes". In it, Arbuthnot examined birth records in London across the years 1629 to 1710. It is one of the earliest documented applications of significance tests. For more historical details, see https://en.wikipedia.org/wiki/History_of_statistics#Origins_in_probability_theory
