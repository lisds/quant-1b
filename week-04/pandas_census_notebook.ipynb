# Pandas Coding Lab: UK Census Data Wrangling

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## 1. Load the dataset

```python
df = pd.read_csv("../data/example_census_dataset.csv")
df.head()
```

## 2. Inspect the dataset

```python
df.shape
df.columns
df.info()
```

## 3. Filter for your chosen area

```python
area_df = df[df["Geography"] == "Leeds"]
area_df.head()
```

## 4. Sort to find top categories

```python
area_df.sort_values("Count", ascending=False).head(5)
```

## Extension: Plotting

```python
top5 = area_df.sort_values("Count", ascending=False).head(5)

plt.bar(top5["Category"], top5["Count"])
plt.xticks(rotation=45)
plt.show()
```
